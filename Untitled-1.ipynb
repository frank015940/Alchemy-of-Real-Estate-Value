* City: The city in which the respective zip code is situated.
* State: The state where the zip code is located.
* Metro: The metropolitan area encompassing the zip code.
* CountyName: The county within which the zip code falls.
* SizeRank: A numerical ranking reflecting the size of the zip code, ranked from 1 to 14723.

```python
# Calculate 4-year ROI for each zip code
df['4_yr_ROI'] = (df['2018-04'] - df['2014-07']) / df['2014-07']

# Identify top 5 zip codes with highest ROI
top_5_zipcodes = df.nlargest(5, '4_yr_ROI')

# Print the top 5 zip codes
print(top_5_zipcodes['RegionName'])
```

```python
# Assuming 'top_5_zipcodes' is your dataframe containing the top 5 zip codes
zip_codes = top_5_zipcodes['RegionName'].values

# Filter the original dataframe to include only the top 5 zip codes
top_5_data = zillow[zillow['RegionName'].isin(zip_codes)]

# Melt the dataframe to make it suitable for time series plotting
top_5_data_melted = top_5_data.melt(id_vars=['RegionName'], var_name='Date', value_name='Value')

# Convert the 'Date' column to datetime format
top_5_data_melted['Date'] = pd.to_datetime(top_5_data_melted['Date'])

# Plot the data
plt.figure(figsize=(10,6))
sns.lineplot(data=top_5_data_melted, x='Date', y='Value', hue='RegionName')
plt.title('Housing Sales Values for Top 5 Zip Codes (1996-2018)')
plt.show()
```
